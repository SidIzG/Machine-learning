```
#Q1
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```
x = x.flatten()
y = y.flatten()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```
eqn = np.poly1d(np.polyfit(x,y,4))
a = np.linspace(-20,5,100)
print(eqn)
```

```
%matplotlib inline

plt.plot(x,y, 'ro', a, eqn(a))
```

```
#Q2:
pred_dict = {}

for n in range(1,9):
  eqn = np.poly1d(np.polyfit(X_train,y_train,n))
  prd = eqn(X_test)
  pred_dict[n] = prd

mse_dict = {}

for n in range(1,9):
  yhat = pred_dict[n]
  mse_dict[n] = (1/len(X_test))*(sum((y_test-yhat)**2))

pred_dict
mse_dict
```

```
%matplotlib inline

m = np.array(list(mse_dict.keys()))
mse = np.array(list(mse_dict.values()))

from matplotlib.pyplot import figure
plt.xlabel('m')
plt.ylabel('MSE')
plt.plot(m, mse, 'rx', m, mse)
```

```
best_m = min(mse_dict, key=mse_dict.get)
best_m
```

```
num_of_samp_dict = {}
```

```
noise_scale_dict = {}
eqn = np.poly1d(np.polyfit(x,y,best_m))
noise_scale_dict[noise_scale] = (1/len(x))*(sum((y-eqn(x))**2))
```
